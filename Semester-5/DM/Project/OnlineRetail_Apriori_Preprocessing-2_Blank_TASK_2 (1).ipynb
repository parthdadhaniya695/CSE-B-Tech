# Step 1: Load Data

```python
import numpy as np
import pandas as pd

df = pd.read_csv('OnlineRetail.csv', encoding='ISO-8859-1') 
print(df)
```

# Step 2: Drop Missing Invoices or Descriptions

```python
df = df.dropna(subset=['InvoiceNo', 'Description'])
df
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
df
```

# Step 4: Keep Only Positive Quantities

```python
df = df[df['Quantity'] > 0]
df
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
df['Description'] = df['Description'].str.lower().str.strip()
df
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
df = df[df['Country'] == 'United Kingdom']
df
```

# Step 7: Remove Duplicates

```python
df = df.drop_duplicates()
df
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df
```

# Step 9 : Check for Top 10 Selling Products

```python
top_products = df.groupby('Description')['Quantity'].sum().nlargest(10)
print(top_products)

import matplotlib.pyplot as plt
top_products.plot(kind='bar', title='Top 10 Selling Products by Quantity')
plt.xlabel('Product Description')
plt.ylabel('Total Quantity Sold')
plt.show()
```

# Step 10 : Identify Most Active Customers

```python
top_customers = df.groupby('CustomerID')['InvoiceNo'].nunique().nlargest(10)
print(top_customers)
```

# Step 11: Time-Based Analysis | Convert InvoiceDate to datetime and check popular months/hours.

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df['Month'] = df['InvoiceDate'].dt.month_name()
df['Hour'] = df['InvoiceDate'].dt.hour

popular_months = df['Month'].value_counts()
print("Popular Months:\n", popular_months)

popular_hours = df['Hour'].value_counts().sort_index()
print("\nPopular Hours:\n", popular_hours)
```
